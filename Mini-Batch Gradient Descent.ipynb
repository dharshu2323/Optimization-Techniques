```python
import numpy as np

# Define the objective function (simple quadratic function)
def objective_function(x):
    return x**2 + 5*x + 6

# Define the gradient of the objective function
def gradient(x):
    return 2*x + 5

# Mini-Batch Gradient Descent function
def mini_batch_gradient_descent(initial_x, learning_rate, batch_size, num_iterations):
    x = initial_x
    data_size = 1000  # Size of the dataset
    
    for iteration in range(num_iterations):
        # Generate a random mini-batch
        indices = np.random.choice(data_size, size=batch_size, replace=False)
        mini_batch = np.array([gradient(idx) for idx in indices])
        
        # Compute the average gradient of the mini-batch
        avg_gradient = np.mean(mini_batch)
        
        # Update the parameter
        x -= learning_rate * avg_gradient
        
        print(f"Iteration {iteration+1}: x = {x:.4f}, Objective = {objective_function(x):.4f}")
    
    return x

# Hyperparameters
initial_x = -10       # Initial guess for x
learning_rate = 0.1   # Learning rate
batch_size = 32       # Size of the mini-batch
num_iterations = 50   # Number of iterations

# Run Mini-Batch Gradient Descent
final_x = mini_batch_gradient_descent(initial_x, learning_rate, batch_size, num_iterations)
print(f"Final x: {final_x:.4f}, Final Objective: {objective_function(final_x):.4f}")
```
